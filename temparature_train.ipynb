## Saving weights of nodes in model

```python
def save_model(array, name):
    np.savetxt(name+".csv", array, delimiter=",")
```

## Loading saved model to predict rainfall

```python
def load_model(name):
    with open(name+".csv", 'r') as read_file:
        reader = csv.reader(read_file)
        data = []
        for row in reader:
            data.append(row[0])
    read_file.close()
    data = np.asarray(data, dtype=np.float64)
    return data
```

## Temprature prediction function by passing model name and year for temp prediction

```python
def _predict_temp(model, year):
    total = 0
    for i in range(1,13):
        x = year+(year*i)/12
        eq_sum = 0
        for j in range(0, model.shape[0]):
            eq_sum += (model[j] * (x**(model.shape[0]-j-1)))
        
        if eq_sum < 0:
            eq_sum = 0
        print(eq_sum)
        total += eq_sum
    avrg = total/12
    return avrg
```
